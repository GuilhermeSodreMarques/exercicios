```python
 EXERCÍCIO 2 - Listagem formatada
# ------------------------------------------------------------
# Imprima cada material no formato:
#   "Cimento CP-II 50kg | Alvenaria | 120 un | R$ 32.50"
```

```python
import csv

with open("materiais.csv", newline="", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    for linha in leitor:
        print(f"{linha['material']} | {linha['categoria']} | {linha['quantidade']} und. | R$: {float(linha['preco_unitario']):.2f}")
```

```python
# EXERCÍCIO 3 - Valor total por item
# ------------------------------------------------------------
# Para cada material, calcule o valor total gasto
# (quantidade * preco_unitario) e imprima:
#   "Cimento CP-II 50kg -> R$ 3900.00"
```

```python
with open ("materiais.csv", newline="", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    for linha in leitor:
        qtd = int(linha["quantidade"])
        preco = float(linha["preco_unitario"])
        total = qtd * preco
        
        print(f"{linha['material']} -> R$ {total:.2f}")        
```

# EXERCÍCIO 4 - Valor total geral
# ------------------------------------------------------------
# Calcule e imprima o valor total gasto somando TODOS os
# materiais (quantidade * preco_unitario de cada um, somados).



```python
total_geral = 0
with open("materiais.csv", newline="", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    for linhas in leitor:
        qtd = int(linha["quantidade"])
        preco = float(linha["preco_unitario"])
        total_geral += qtd * preco
print(f"valor total gasto: R$ {total_geral:.2f}")
```

 EXERCÍCIO 6 - Maior e menor valor
# ------------------------------------------------------------
# Descubra e imprima:
#   a) o material com o MAIOR valor total (quantidade * preco)
#   b) o material com o MENOR valor total


```python
materiais_com_total =[] #vai guardar tuplas

with open("materiais.csv", newline="", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    for linha in leitor:
        qtd = int(linha["quantidade"])
        preco = float(linha["preco_unitario"])
        total = qtd * preco
        materiais_com_total.append((linha["material"], total))
        
maior = max(materiais_com_total, key=lambda item: item [1])
menor = min(materiais_com_total, key=lambda item: item[1])

print(f"Maior valor total: {maior[0]}  -> R$ {maior[1]:.2f}")
print(f"Menor valor total: {menor[0]}  -> R$ {menor[1]:.2f}")
```

EXERCÍCIO 7 - Agrupamento (o mais avançado desta lista)
# ------------------------------------------------------------
# Calcule o valor total gasto POR CATEGORIA e imprima algo como:
#   Alvenaria: R$ 8437.50
#   Acabamento: R$ ...
#   Eletrica: R$ ...
#   Hidraulica: R$ ...
#   Cobertura: R$ ...

```python
totais_por_categoria = {}

with open("materiais.csv", newline="", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    for linha in leitor:
        categoria = linha["categoria"]
        qtd = int(linha["quantidade"])
        preco = float(linha["preco_unitario"])
        total = qtd * preco
        
        totais_por_categoria[categoria] = totais_por_categoria.get(categoria,0) + total #o .get(categoria,0) retorna o valor atual da categoria.
        
    for categoria, total in totais_por_categoria.items():
        print(f"{categoria}: R$ {total:.2f}")
```
